```python
import matplotlib as plot

c2 = pd.read_csv("datasets/covid-19-pandemic-worldwide-data.csv", delimiter = ';') 

df2 = pd.DataFrame(c2)
df2
```

```python
df_mask = df2['Category'] == 'Recovered'
recovered_df = df2[df_mask].groupby(['Zone'])['Count'].max().sort_values(ascending = False)[:10]
recovered_df.plot.bar()
```

```python
df_mask = df2['Category'] == 'Deaths'
deaths_df = df2[df_mask].groupby(['Zone'])['Count'].max().sort_values(ascending = False)
deaths_df

```

```python
res_pos = (recovered_df / deaths_df).sort_values(ascending = False)[:10]
res_pos.replace([np.inf, -np.inf], np.nan).dropna()
```

```python
res_neg = (deaths_df / recovered_df).sort_values(ascending = False)[:10]
res_neg.replace([np.inf, -np.inf], np.nan).dropna()
```
